prac 12:student record mgmt dict and functions

```python
student={}
def add_stud():
    name=input("Enter stud name:")
    marks=int(input("Enter stud marks:"))
    student[name]=marks
    print("record added ")

def search_stud():
    name=input("Enter name of stud to be search:")
    if name in student :
        print(f"{name}'s marks : {student[name]}\n")
    else:
        print("no record")

def del_stud():
    name=input("Enter name of stud to be delete:")
    if name in student:
        del student[name]
        print("record deleted")
    else:
        print("no record")

def highest():
    if student:
        name = max(student, key=student.get)
        print(f"{name} having highest marks: {student[name]}")
    else:
        print("No records available")

def display():
    if student:
        print("\n ------stud RECORDS-----")
        for name,marks in student.items():
            print(f"{name}:{marks}")
            print()  
    else:
        print("NO record available")

while True:
    print("1. Add stud")
    print("2. Search stud")
    print("3. Delete stud")
    print("4. Highest marks")
    print("5. Display stud")
    print("6. EXIT")
    
    choice = input("Enter your choice(1-6):")
    if choice == "1":
        add_stud()
    elif choice == "2" :
        search_stud()
    elif choice == "3" :
        del_stud()
    elif choice == "4":
        highest()
    elif choice == "5":
        display()
    elif choice == "6" :
        print("Exiting from program")
        break
    else:
        print("Invalid choice")
```
