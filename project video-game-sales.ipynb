## **Segment data by region and compare sales trends across different geographical areas**

#### Create a subset for each region

```python
#segment data by region
na_sales = df.groupby('Year')['NA_Sales'].sum().reset_index()
eu_sales = df.groupby('Year')['EU_Sales'].sum().reset_index()
jp_sales = df.groupby('Year')['JP_Sales'].sum().reset_index()
other_sales = df.groupby('Year')['Other_Sales'].sum().reset_index()
```

#### Comparing by bar plots

```python
# Merge data into a single DataFrame for easier plotting
sales = na_sales.merge(eu_sales, on='Year').merge(jp_sales, on='Year').merge(other_sales, on='Year')
sales.set_index('Year', inplace=True)

# Plot sales trends using bar plots
sales.plot(kind='bar', figsize=(14, 8))

plt.title('Sales Trends by Region')
plt.xlabel('Year')
plt.ylabel('Total Sales (in millions)')
plt.legend(title='Region')
plt.show()
```

#### Comparing by boxplot

```python
# Melt the DataFrame to long format for seaborn boxplot
melted_sales = sales.reset_index().melt(id_vars='Year', var_name='Region', value_name='Sales')

# Plot box plots
plt.figure(figsize=(14, 8))
sns.boxplot(x='Region', y='Sales', data=melted_sales)

plt.title('Sales Distribution by Region')
plt.xlabel('Region')
plt.ylabel('Sales (in millions)')
plt.show()
```

## **Visualizations**

### **Bar Charts**

#### Bar chart for top 10 Publishers 

```python
publisher_sales = df.groupby('Publisher')['Global_Sales'].sum().reset_index()
# Sort the publishers by total global sales and select the top 10
top_publishers = publisher_sales.sort_values(by='Global_Sales', ascending=False).head(10)

# Create the bar chart
plt.figure(figsize=(14, 7))
plt.bar(top_publishers['Publisher'], top_publishers['Global_Sales'], color='skyblue')
plt.xlabel('Publisher')
plt.ylabel('Total Global Sales (in millions)')
plt.title('Top 10 Publishers by Global Sales')
plt.xticks(rotation=45)
plt.show()
```

#### Bar Chart for Sales Distribution by Genre

```python
genre_sales = df.groupby('Genre').sum()
plt.figure(figsize=(12, 6))
genre_sales['Global_Sales'].plot(kind='bar', color='Pink')
plt.xlabel('Genre')
plt.ylabel('Global Sales (in millions)')
plt.title('Sales Distribution by Genre')
plt.xticks(rotation=45)
plt.show()
```

#### Bar chart for top 10 Games by Global Sales

```python
# Sort games by global sales to find out th top 10
top_games = df.sort_values('Global_Sales', ascending=False).head(10)

# Plot top 10 games by global sales
plt.figure(figsize=(14, 7))
plt.barh(top_games['Name'], top_games['Global_Sales'], color='lightcoral')
plt.xlabel('Global Sales (in millions)')
plt.ylabel('Game Title')
plt.title('Top 10 Games by Global Sales')
#plt.gca().invert_yaxis()
plt.show()
```

####  Bar chart for Average Sales per year

```python
# Group by Year and Genre and then calculate the mean sales
avg_sales_by_year_genre = df.groupby(['Year', 'Genre'])['Global_Sales'].mean().unstack()

# Plot average sales per year by genre
avg_sales_by_year_genre.plot(kind='bar', stacked=True, figsize=(14, 7))
plt.xlabel('Year')
plt.ylabel('Average Global Sales (in millions)')
plt.title('Average Sales per Year by Genre')
plt.legend(title='Genre')
plt.show()
```

#### Bar Chart for Sales distribution 

```python
# Aggregate sales by platform
platform_sales = df.groupby('Platform').sum()

# Plot sales distribution by platform
plt.figure(figsize=(14, 7))
platform_sales['Global_Sales'].sort_values(ascending=False).plot(kind='bar', color='orange')
plt.xlabel('Platform')
plt.ylabel('Global Sales (in millions)')
plt.title('Sales Distribution by Platform')
plt.xticks(rotation=45)
plt.show()
```

### **Pie charts**

#### Pie chart for Sales Distribution by region


```python
# Aggregate sales by region
region_sales = df[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']].sum()

# Plot pie chart for sales distribution by region
plt.figure(figsize=(10, 6))
plt.pie(region_sales, labels=region_sales.index, autopct='%1.1f%%', startangle=140, colors=['lightblue', 'lightgreen', 'lightcoral', 'lightskyblue'])
plt.title('Sales Distribution by Region')
plt.axis('equal')  # Equal ensures that pie is drawn as a circle.
plt.show()
```

#### Pie chart for Sales distribution by Genre

```python
# Aggregate sales by genre
genre_sales = df.groupby('Genre')['Global_Sales'].sum()

# Plot pie chart for sales distribution by genre
plt.figure(figsize=(12, 7))
plt.pie(genre_sales, labels=genre_sales.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired(range(len(genre_sales))))
plt.title('Sales Distribution by Genre')
plt.axis('equal')  # Equal  ensures that pie is drawn as a circle.
plt.show()
```

#### Pie chart for Sales Distribution by Platform
